```python
import numpy as np
import cv2 as cv
from tensorflow.python.keras.models import model_from_json
```

```python
#загружаем нейронную сеть
json_file = open("model_gen_j_99.json", "r")
loaded_model_json = json_file.read()
json_file.close()

model = model_from_json(loaded_model_json)
model.load_weights("model_gen_h_99.h5")

#компилиркем нейроннуюсеть
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
cap = cv.VideoCapture(0)
list_class = ['a_unevenness/Selection_139.jpg',
                  'no_drive/1932.png',
                  'no_entry/40.jpg',
                  'none/29.png',
                  'parking/67.jpg',
                  'pedistrain/66.jpg',
                  'road_works/127.jpg',
                  'stop/396.png',
                  'way_out/(76).jpg'
                 ]
def color_filter(low_HSV, high_HSV, frame):
        hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
        lower_hsv = np.array(low_HSV, dtype = "uint8")
        higher_hsv = np.array(high_HSV, dtype = "uint8")
        mask = cv.inRange(hsv, lower_hsv, higher_hsv)
        return mask
    
def get_masks(im):
        # создает маску для синего красного и черного
        red_mask = color_filter([0,189,150], [179,255,255], im)
        # black_mask = color_filter([60,50,0], [143,200,255], im)
        blue_mask = color_filter([70,50,0], [129,255,250], im)
        red_mask = red_mask + blue_mask
        red_mask = cv.dilate(red_mask, None, iterations = 3)
        red_mask = cv.erode(red_mask, None, iterations = 2)
        return red_mask
    
while True:
    ret, frame = cap.read()
    frame = cv.resize(frame, (320,240))
    cv.imshow('Frame', frame)
    frame_bin = get_masks(frame)
    frame_draw = frame.copy()
    contur = cv.findContours(frame_bin, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)

    contur = contur[1]
    if contur:
        contur = sorted(contur, key=cv.contourArea, reverse=True)
        #cv2.imshow('test_new', frame_new)
        cv.drawContours(frame_draw, contur[0], 0, (225, 0, 225), 3)
        # тут начинается вырезание
        (x, y, w, h) = cv.boundingRect(contur[0])
        cv.rectangle(frame, (x, y), (x + w, y + h), (0, 225, 0), 2)
        frame = frame[y:y + h, x:x + w]
        #cv2.imshow('little', frame_new)
    
   
    #подготовка изображения к распознованию
    frame = cv.resize(frame, (80, 80))
    frame = 255 - frame
    frame = frame / 255
    frame = np.expand_dims(frame, axis=0)
    
    prediction = model.predict(frame)
    
    Dir = 'data/training/' + list_class[np.argmax(prediction)]
    result = cv.imread(Dir)
    result = cv.resize(result, (320,240))
    #frame = cv.resize(frame,(320,240))
    #cv.imshow('fr', frame)
    cv.imshow('bin', frame_bin)
    cv.imshow('contr', frame_draw)
    cv.imshow('Result', result)
    
    print(np.argmax(prediction))
    
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
#cv.imwrite('test_img.png', frame)
```
